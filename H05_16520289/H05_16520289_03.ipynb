```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nilai_uas.csv")
df.head()
```

```python
df["nilai_kal"].plot(kind="hist",rwidth=0.8,title="Sebaran nilai kalkulus")
plt.show()
```

```python
# Grafik diatas adalah histogram sebaran nilai kalkulus.
```

```python
df_fak = pd.DataFrame({"frequency": df["fakultas"].value_counts()})
df_fak.plot(kind="barh",y="frequency",title="Sebaran fakultas")
plt.show()
```

```python
# <!-- Chart di atas adalah horizontal chart dari jumlah siswa tiap fakultas. -->
```

```python
df_fak.plot(kind="pie",y="frequency",legend=None,title="Sebaran mahasiswa tiap fakultas",figsize=(12,6),autopct='%.2f%%')
plt.ylabel('')
plt.show()
```

```python
# Fakultas dengan mahasiswa terbanyak adalah STEI.
# Diagram yang lebih baik untuk menunjukkan fakultas dengan mahasiswa terbanyak adalah chart horizontal, karena dapat terlihat dengan lebih jelas yang mana yang lebih tinggi.
```

```python
df.plot(kind="scatter",x="nilai_kim",y="nilai_fis",title="Sebaran nilai fisika dan kimia")
plt.show()
```
