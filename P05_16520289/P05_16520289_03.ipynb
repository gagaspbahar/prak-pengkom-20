```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("crypto.csv")
```

```python
# Nomor 1, Line chart untuk ETH dengan nilai Close.
```

```python
df.loc[df["name"] == "XMR"].plot(kind="line", x="date",y="close",figsize=(12,5))
plt.show()
```

```python
# Nomor 2, Scatter plot high-low untuk USDT
```

```python
df.loc[df["name"] == "USDT"].plot(kind="scatter",x="high",y="low",title="Scatterplot high dan low koin USDT")
plt.show()
```

```python
# Nomor 3, Pie chart banyak data tiap koin.
```

```python
df["name"].value_counts().plot(kind="pie",title="Banyak data tiap coin")
plt.ylabel("")
plt.show()
```

```python
# Nomor 4, Pie chart dari Market pada tanggal 2 Maret 2015
```

```python
df.loc[df["date"]=="2015-03-02"].set_index("name")["market"].plot(kind="pie",legend=1,title="Pie chart nilai kapitalisasi pasar",figsize=(12,5),autopct='%.2f%%')
plt.ylabel("")
plt.show()
```

```python
# Nomor 5, Line chart close dari semua koin.
```

```python
df.groupby(["date","name"])["close"].sum().unstack().plot(kind="line",figsize=(13,5),title="Perbandingan nilai close setiap koin")
plt.xlabel("Tanggal")
plt.ylabel("Close")
plt.show()
```
